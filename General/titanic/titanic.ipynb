```python
def clasificador_titanic(x):
    vive = False
    if x['Age'] < 18:
        vive = True
    if x['Sex'] == "female" and x['Pclass'] == 1:
        vive = True

    if x['Sex'] == "female" and  0 < x['SibSp']:
        vive = True
    
    if x['Sex'] == "male" and x['SibSp'] == 0 and x['Pclass'] == 3:
        vive = False
    
    
    
    return vive



    
```

```python
def clasificador_titanic(x):
    # Define rules for survival based on combinations of Sex, Pclass, Age, and SibSp
    rules = {
        # genero, clase, edad, hermanos
        ('female', 1, True, True): True,
        ('female', 1, True, False): True,
        ('female', 1, False, True): True,
        ('female', 1, False, False): True,
        ('female', 2, True, True): True,
        ('female', 2, True, False): True,
        ('female', 2, False, True): True,
        ('female', 2, False, False): False,
        ('female', 3, True, True): True,
        ('female', 3, True, False): True,
        ('female', 3, False, True): False,
        ('female', 3, False, False): False,
        ('male', 1, True, True): True,
        ('male', 1, True, False): True,
        ('male', 1, False, True): True,
        ('male', 1, False, False): False,
        ('male', 2, True, True): False,
        ('male', 2, True, False): False,
        ('male', 2, False, True): False,
        ('male', 2, False, False): False,
        ('male', 3, True, True): False,
        ('male', 3, True, False): False,
        ('male', 3, False, True): False,
        ('male', 3, False, False): False,
    }
    
    # Determine the combination
    age_group = x['Age'] < 18
    sibsp_group = x['SibSp'] >= 1
    combination = (x['Sex'], x['Pclass'], age_group, sibsp_group)
    
    # Return the survival decision based on the rules
    return rules.get(combination, False)

# implement this function to the dataframe with a for
df_titanic['Survived_prediction'] = df_titanic.apply(clasificador_titanic, axis=1)

#test the accuracy of the model 
from sklearn.metrics import accuracy_score
restult = accuracy_score(df_titanic['Survived'], df_titanic['Survived_prediction'])
restult
```
